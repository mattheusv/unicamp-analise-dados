```python
import pandas
```

```python
cpu = pandas.read_csv('cpu.csv')
#memory = pandas.read_csv('memory.usage.percentage_23-05-27.csv')
#sales = pandas.read_csv('sales.total_23-05-27.csv')

cpu.index = pandas.to_datetime(cpu['timestamp'], format='ISO8601')
#memory.index = pandas.to_datetime(cpu['timestamp'], format='ISO8601')
#sales.index = pandas.to_datetime(cpu['timestamp'], format='ISO8601')

del cpu['timestamp']
# del cpu['ts_name']
#del memory['timestamp']
#del sales['timestamp']
```

```python
cpu_avg = cpu.resample('M').median()
cpu_avg
```

```python
from sklearn import linear_model
```

```python
x=cpu_avg.index
y=cpu_avg['value'].values.reshape(-1, 1)

lm = linear_model.LinearRegression()
model = lm.fit(x.values.reshape(-1, 1), y)

predictions = lm.predict(x.values.astype(float).reshape(-1, 1))


cpu_avg['predictions'] = predictions
cpu_avg.plot(style={'bat': 'or'})
pyplot.ylabel('value')
pyplot.legend()
```
