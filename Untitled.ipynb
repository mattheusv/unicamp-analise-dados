```python
import pandas
import matplotlib.pyplot as pyplot
import seaborn
from sklearn import linear_model
```

```python
cpu = pandas.read_csv('cpu.csv')

cpu.index = pandas.to_datetime(cpu['timestamp'], format='ISO8601')

del cpu['timestamp']
```

```python
cpu_avg = cpu.resample('M').median()
cpu_avg
```

```python
# seaborn.set()
colors = [
    '#000000',  # teal/cyan
    '#FE53BB',  # pink
    '#F5D300',  # yellow
    '#00ff41'  # matrix green
             ]

cpu_avg.plot(marker='o', color=colors)
```

```python
import pandas
from sklearn.linear_model import LinearRegression
df = pandas.read_csv('cpu.csv')

df['timestamp'] = pandas.to_datetime(df['timestamp'])
#df['value'] = df['value'].astype(float)

df.set_index('timestamp', inplace=True)
df = df.resample('M').median()

df2 = df.copy()
df2['timestamp'] = df2.index.view('int64')
df2.set_index('timestamp', inplace=True)


X = df2.index.values.reshape(-1, 1)
y = df2['value'].values


model = LinearRegression()
model.fit(X, y)

# Generate timestamps for the next 3 months
last_date = df2.index[-1]
next_3_months_num = pandas.date_range(start=last_date, periods=6, freq='M').view('int64')
predictions = model.predict(next_3_months_num.reshape(-1, 1))

next_3_months = pandas.to_datetime(next_3_months_num)

forecast_df = pandas.DataFrame({'date': next_3_months, 'prediction': predictions})

forecast_df.set_index('date', inplace=True)

combined_df = pandas.concat([df, forecast_df])
combined_df

combined_df.plot(marker='o')
```
